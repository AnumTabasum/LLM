```python
import torch
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Words in the sentence
words = ["dream", "big", "and", "work", "for", "it"]

# Manually created 3D embeddings (example)
embeddings = torch.tensor([
    [0.2, 0.8, 0.5],   # dream
    [0.9, 0.4, 0.6],   # big
    [0.1, 0.3, 0.2],   # and
    [0.7, 0.9, 0.4],   # work
    [0.4, 0.2, 0.7],   # for
    [0.6, 0.5, 0.9]    # it
])

print(embeddings)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = embeddings[:, 0]
y = embeddings[:, 1]
z = embeddings[:, 2]

ax.scatter(x, y, z)

# Label each point with the word
for i, word in enumerate(words):
    ax.text(x[i], y[i], z[i], word)

ax.set_xlabel("Dimension 1")
ax.set_ylabel("Dimension 2")
ax.set_zlabel("Dimension 3")

plt.show()
```
